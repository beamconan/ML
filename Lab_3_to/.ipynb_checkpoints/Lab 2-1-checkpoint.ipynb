# ตอนที่ 1: การทดลองเตรียมข้อมูล ปรับค่าข้อมูล และจัดแบ่งชุด Train, Test เพื่อสอนโมเดล

1.1 เตรียมข้อมูลทดลอง

```python
# Stock data
import quandl
import datetime
# Analyzing
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn import model_selection
from sklearn import preprocessing
from sklearn import metrics
from scipy import stats
```

```python
# load data to csv
# data = quandl.get("WIKI/AMZN")
# data = data.to_csv('data.csv')
```

```python
# Read file csv
data = pd.read_csv('data.csv')
data
```

1.2 ปรับรูปแบบของข้อมูล

- เลือกใช้ข้อมูล ‘Close’ ซึ่งเป็นราคาปิดหรือราคาสุดท้ายของแต่ละวันมาใช้

```python
df= pd.DataFrame(data,columns=['Date','Close'])
df['Date'] = pd.to_datetime(df['Date'])
df.sort_values('Date', inplace=True)
df
```

- สร้างข้อมูลทางเลือกด้วยการทำ Normalization ข้อมูล ‘Close’ ที่เลือก

    - Data cleaning

```python
df[df.columns[1:]] = df[df.columns[1:]].apply(pd.to_numeric, errors='coerce')
df = df.fillna(df.median()) # จัดการข้อมูลหาย แทนที่ด้วย median
df = df.drop_duplicates()
```

        - Visualize Data

```python
hist = df.hist(bins=8,figsize=(3,3))
```

```python
axes = df.plot.line(x='Date', subplots=True, figsize=(5, 5))
```

        - หา timediff แล้วทำการ resample

```python
df['timediff'] = (df['Date']-df['Date'].shift()).fillna(0)
print('median timediff : ',df['timediff'].median())
df = df.set_index('Date').resample('1D').mean()
df
```

```python
df.interpolate(inplace=True)
df
```

```python
df.plot.line(subplots=True, figsize=(5, 5))
```

    - Data Normalization (Standardized Norm)

```python
x = df.values #returns a numpy array
standard_scaler = preprocessing.StandardScaler() #Standardized Norm
x_standard_scaled = standard_scaler.fit_transform(x)
df = pd.DataFrame(x_standard_scaled,columns=['Close'])
df.plot()
```

        - สร้างข้อมูลราคาวันถัดไป Next_N-day เพื่อใช้เป็นคำตอบ (Ground Truth) ในการคาดการณ์ (predict ข้อมูลในอีกหลายวันข้างหน้า กำหนดวันในการคาดการณ์จำนวน Next_N-day วันล่วงหน้า

```python
Next_N_day = 30 # predict ข้อมูลอีก 30 วันข้างหน้า
Close = df.iloc[:-Next_N_day]
GT = df.iloc[Next_N_day:]
```

1.3 จัดเตรียมข้อมูลสำหรับ train validation และ test

- ข้อมูล test ให้ใช้ข้อมูล ‘GT’ ช่วง 60 วันท้าย ซึ่งตรงกับข้อมูล ‘Close’ ตัดจำนวนวันคาดการณ์ล่วงหน้า (Next_N-day) ไปแล้ว 60 วันท้าย

```python
# test set
X_test = Close.iloc[-60:]
y_test = GT.iloc[-60:]
```

- เตรียมข้อมูล train, validate ด้วยการใช้ฟังก์ชัน model_selection.train_test_split() กำหนด random_seed เพื่อให้ผลการจัดข้อมูลคงที่เหมือนกันทุกครั้งที่สอนโมเดล

```python
# training set and validation set
seed = 42
X_train, X_validate, y_train, y_validate = model_selection.train_test_split(Close.iloc[:-60], GT.iloc[:-60], test_size=0.33, random_state=seed)
```

1.4 แสดงรูปกราฟการกระจายของ train validate ที่แบ่งจากข้อ 1.3

```python
alpha = 0.3

fig = plt.figure(figsize=(30,10))
ax = fig.add_subplot(111)

ax.scatter(X_train, y_train, marker='o', color='blue',alpha=alpha,label="Training set");
ax.scatter(X_validate, y_validate, marker='o', color='pink',alpha=alpha,label="Test set");

ax.grid(True)
plt.legend()
plt.title("Training set")
plt.show
```
