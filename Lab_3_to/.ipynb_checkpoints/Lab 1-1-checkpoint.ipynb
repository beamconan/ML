# ตอนที่ 1: การทดลองอ่านไฟล์ข้อมูล การแก้ปัญหา ข้อมูลหาย และการปรับช่วงค่าของข้อมูล แสดงข้อมูลเชิงกราฟ และ การจัดเตรียมรูปแบบข้อมูลเพื่อนำเข้าโมเดล

1.1 ขั้นตอนการทดลองในการนำเข้าข้อมูล แก้ปัญหาข้อผิดพลาดของข้อมูล และการปรับช่วงค่าของข้อมูล

- Import Library

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from scipy import stats
from mpl_toolkits.mplot3d import axes3d, Axes3D

%matplotlib inline
```

- Read Data File

```python
df = pd.read_csv('watch_test2_sample.csv')
df['uts'] = pd.to_datetime(df['uts'])
df.sort_values('uts', inplace=True)
_df = df.copy()
df
```

- Data Cleaning

(1) จัดการข้อมูลหาย หรือ ซ้ำซ้อน โดยสามารถเลือกตัดทิ้ง หรือ แทนที่ด้วย 0, mean, median [ pd.fillna() , pd. replace() , pd. drop_duplicates() ]

```python
df[df.columns[1:]] = df[df.columns[1:]].apply(pd.to_numeric, errors='coerce')
df_missing_zero = df.fillna(0) # จัดการข้อมูลหาย แทนที่ด้วย 0
df_missing_median = df.fillna(df.median()) # จัดการข้อมูลหาย แทนที่ด้วย median
df_missing_mean = df.fillna(df.mean()) # จัดการข้อมูลหาย แทนที่ด้วย mean
```

```python
df_missing = df_missing_median # เลือกใช้ การจัดการข้อมูล โดยแทนที่ด้วย median
df_missing.shape
```

```python
df_drop_dup = df_missing.drop_duplicates() # จัดการข้อมูลซ้ำซ้อน โดย ตัดทิ้ง 
df_drop_dup
```

    - Visualize Data

```python
hist = df_drop_dup.hist(bins=8,figsize=(10,10))
```

```python
axes = df_drop_dup.plot.line(x='uts', subplots=True, figsize=(8, 36))
```

            - remove outlier

```python
# Remove outlier
list_remove_outlier = ['compass','heartrate','light','pressure']
threshold = 3
for li in list_remove_outlier:
    df_drop_dup.iloc[np.abs(stats.zscore(df_drop_dup[li])) > 3,df_drop_dup.columns.get_loc(li)] = df_drop_dup[li].median()
```

```python
axes = df_drop_dup.plot.line(x='uts', subplots=True, figsize=(8, 36))
```

(2) แทรกข้อมูลในช่วงเวลาที่ขาดหายไป [pd.interpolate()]

    - หา timediff

```python
df_drop_dup['timediff'] = (df_drop_dup['uts']-df_drop_dup['uts'].shift()).fillna(0)
print('median timediff : ',df_drop_dup['timediff'].median())
df_drop_dup
```

    - resample

```python
df_interpol = df_drop_dup.set_index('uts').resample('21s').mean()
df_interpol
```

```python
df_interpol.interpolate(inplace=True)
df_interpol
```

```python
df_interpol.plot.line(subplots=True, figsize=(8, 36))
```

(3) จัดการลดสัญญาณรบกวนในข้อมูลด้วยการทำ Moving Average [ pd.rolling() ]

```python
df_cleaning = df_interpol.rolling(5).mean()
df_cleaning
```

Data Normalization

```python
x = df_cleaning.values #returns a numpy array
```

```python
min_max_scaler = preprocessing.MinMaxScaler() # Max-Min Norm
x_min_max_scaled = min_max_scaler.fit_transform(x)
df_min_max = pd.DataFrame(x_min_max_scaled)
df_min_max.plot()
```

```python
standard_scaler = preprocessing.StandardScaler() #Standardized Norm
x_standard_scaled = standard_scaler.fit_transform(x)
df_standard = pd.DataFrame(x_standard_scaled)
df_standard.plot()
```

```python
df = df_standard # เลือกใช้ Standardized Norm
df
```

1.3 ขั้นตอนการจัดเตรียมข้อมูลเพื่อนำเข้าโมเดล

- ทำการจัดข้อมูล 5 Features [accelerateX, accelerateY, accelerateZ, compass, heartrate] ในรูปของอะเรย์ 2 มิติ 2 ชุด

    - ชุดที่ 1 

```python
columns = [0,1,2,3,9]
```

```python
array_1 = df.iloc[4:,columns].to_numpy()
array_1.shape
```

```python
sns.heatmap(array_1,xticklabels=['accelerateX', 'accelerateY', 'accelerateZ', 'compass', 'heartrate'] ,  annot=False)
```

    - ชุดที่ 2

```python
timestep = 4
stride = 2
data = []
for i in range(4,len(df)-timestep+1,stride):
    data.append(df[columns].iloc[i:i+timestep].to_numpy())

array_2 = np.array(data)
array_2.shape
```

```python
array_2
```

```python
arr_2 = np.concatenate(array_2)
sns.heatmap(arr_2, xticklabels=['accelerateX', 'accelerateY', 'accelerateZ', 'compass', 'heartrate'] ,  annot=False)
```
